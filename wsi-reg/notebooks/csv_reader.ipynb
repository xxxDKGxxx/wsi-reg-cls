```python
import pandas as pd
from IPython.display import display

cols = ['Metryka', 'Mean', 'Std']

def merge_results(simple, cat, num, full, target_ohe):
    s = simple.rename(columns={'Mean': 'Simple_Mean', 'Std': 'Simple_Std'})
    c = cat.rename(columns={'Mean': 'Cat_Mean', 'Std': 'Cat_Std'})
    n = num.rename(columns={'Mean': 'Num_Mean', 'Std': 'Num_Std'})
    f = full.rename(columns={'Mean': 'Full_Mean', 'Std': 'Full_Std'})
    t = target_ohe.rename(columns={'Mean': 'TargetOHE_Mean', 'Std': 'TargetOHE_Std'})
    
    merged = s.merge(c, on='Metryka').merge(n, on='Metryka').merge(f, on='Metryka').merge(t, on='Metryka')
    
    cols_order = [
        'Metryka', 
        'Simple_Mean', 'Cat_Mean', 'Num_Mean', 'Full_Mean', 'TargetOHE_Mean',
        'Simple_Std', 'Cat_Std', 'Num_Std', 'Full_Std', 'TargetOHE_Std'
    ]
    
    return merged[cols_order]

xgb_simple = catalog.load("cv_results_xgboost_simple")[cols]
xgb_cat = catalog.load("cv_results_xgboost_categories")[cols]
xgb_num = catalog.load("cv_results_xgboost_numerical")[cols]
xgb_full = catalog.load("cv_results_xgboost_full")[cols]
xgb_target_ohe = catalog.load("cv_results_xgboost_target_ohe")[cols]

rf_simple = catalog.load("cv_results_rf_simple")[cols]
rf_cat = catalog.load("cv_results_rf_categories")[cols]
rf_num = catalog.load("cv_results_rf_numerical")[cols]
rf_full = catalog.load("cv_results_rf_full")[cols]
rf_target_ohe = catalog.load("cv_results_rf_target_ohe")[cols]

lr_simple = catalog.load("cv_results_lr_simple")[cols]
lr_cat = catalog.load("cv_results_lr_categories")[cols]
lr_num = catalog.load("cv_results_lr_numerical")[cols]
lr_full = catalog.load("cv_results_lr_full")[cols]
lr_target_ohe = catalog.load("cv_results_lr_target_ohe")[cols]

print("XGBoost:")
display(merge_results(xgb_simple, xgb_cat, xgb_num, xgb_full, xgb_target_ohe))

print("\nRandom Forest:")
display(merge_results(rf_simple, rf_cat, rf_num, rf_full, rf_target_ohe))

print("\nLinear Regression:")
display(merge_results(lr_simple, lr_cat, lr_num, lr_full, lr_target_ohe))
```
